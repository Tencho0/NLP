```python
import pandas as pd
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
import re
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------
# 1. Load training data
# -------------------------
df = pd.read_csv("train_addresses.csv")

X = []
y = []

STREET_TYPES = {"ул", "бул", "пл"}
CITIES = {"София", "Пловдив", "Варна", "Бургас"}

def token_features(token, position, total_tokens):
    return {
        "token": token.lower(),
        "is_digit": token.isdigit(),
        "is_upper": token[0].isupper(),
        "position": position,
        "is_first": position == 0,
        "is_last": position == total_tokens - 1,
        "length": len(token),
    }

# -------------------------
# 2. Build token-level dataset
# -------------------------
for _, row in df.iterrows():
    tokens = row["address"].split()
    labels = row["labels"].split()

    for i, (token, label) in enumerate(zip(tokens, labels)):
        X.append(token_features(token, i, len(tokens)))
        y.append(label)

# -------------------------
# 3. Train model
# -------------------------
model = Pipeline([
    ("vectorizer", DictVectorizer()),
    ("classifier", LogisticRegression(max_iter=300))
])

model.fit(X, y)


# --- карта за разширяване на съкращения ---
STREET_MAPPING = {
    "ул": "улица",
    "бул": "булевард",
    "пл": "площад"
}

STREET_TYPES_FULL = {"улица", "булевард", "площад"}

# --- карта за трансформация на латиница към кирилица (основни букви) ---
LATIN_TO_CYRILLIC = {
    "A": "А", "B": "Б", "C": "Ц", "D": "Д", "E": "Е", "F": "Ф",
    "G": "Г", "H": "Х", "I": "И", "J": "Й", "K": "К", "L": "Л",
    "M": "М", "N": "Н", "O": "О", "P": "П", "Q": "К", "R": "Р",
    "S": "С", "T": "Т", "U": "У", "V": "В", "W": "У", "X": "Х",
    "Y": "Й", "Z": "З",
    "a": "а", "b": "б", "c": "ц", "d": "д", "e": "е", "f": "ф",
    "g": "г", "h": "х", "i": "и", "j": "й", "k": "к", "l": "л",
    "m": "м", "n": "н", "o": "о", "p": "п", "q": "к", "r": "р",
    "s": "с", "t": "т", "u": "у", "v": "в", "w": "у", "x": "х",
    "y": "й", "z": "з"
}

def clean_address(address):
    # --- премахване на нелогични символи (оставяме букви, цифри, интервали) ---
    address = re.sub(r"[^a-zA-Z0-9а-яА-Я\s]", "", address)

    # --- латиница към кирилица ---
    address = "".join(LATIN_TO_CYRILLIC.get(c, c) for c in address)

    # --- замяна на съкращения ---
    tokens = address.split()
    tokens = [STREET_MAPPING.get(t, t) for t in tokens]

    # --- съставяме обратно в низ ---
    return " ".join(tokens)
# -------------------------
# 4. Address parser (RULE + ML)
# -------------------------
def parse_address(address):
    tokens = address.split()
    features = [token_features(t, i, len(tokens)) for i, t in enumerate(tokens)]
    predictions = model.predict(features)

    result = {
        "street_type": [],
        "street": [],
        "number": [],
        "city": []
    }

    for token, label in zip(tokens, predictions):
        # --- hard rules override ML ---
        if token.isdigit():
            label = "NUMBER"
        elif token in STREET_TYPES_FULL:
            label = "STREET_TYPE"
        elif token in CITIES:
            label = "CITY"

        if label == "STREET_TYPE":
            result["street_type"].append(token)
        elif label == "STREET":
            result["street"].append(token)
        elif label == "NUMBER":
            result["number"].append(token)
        elif label == "CITY":
            result["city"].append(token)

    return {
        k: " ".join(v) if v else None
        for k, v in result.items()
    }

# -------------------------
# 5. Test
# -------------------------
tests = [
    "ул Средец& 15 София",
    "бул Княз Александър 10 София",
    "ул Иван Петров 5 Пловдив",
    "бул Царство 1 99 София",
    "ул Georgi Rakovski 22 Calapica"
]
final_df = pd.DataFrame()
test_df = pd.DataFrame()
for t in tests:
    cleaned = clean_address(t)
    parsed = parse_address(cleaned)
    print("Original:", t)
    print("Cleaned:", cleaned)
    print("Parsed:", parsed)
    print("-" * 40)
    final_df = pd.concat([final_df, pd.DataFrame([parsed])], ignore_index=True)
    test_df = pd.concat([test_df, pd.DataFrame([{"raw_address": t, "cleaned_address": cleaned}])], ignore_index=True)
display(test_df)
print()
print("adresses")
display(final_df)



plt.figure(figsize=(6,4))
sns.countplot(data=final_df, x="city", palette="muted")
plt.title("Брой адреси по град")
plt.ylabel("Брой")
plt.xlabel("Град")
plt.show()


plt.figure(figsize=(6,4))
sns.countplot(data=final_df, x="street_type", palette="muted")
plt.title("Брой адреси по тип улица")
plt.ylabel("Брой")
plt.xlabel("Тип улица")
plt.show()


plt.figure(figsize=(10,4))
sns.countplot(data=final_df, x="street", palette="muted")
plt.title("Брой имена на улици")
plt.ylabel("Брой")
plt.xlabel("Име на улица")
plt.show()

plt.figure(figsize=(10,4))
sns.countplot(data=final_df, x="number", palette="muted")
plt.title("Брой номера на улици")
plt.ylabel("Брой")
plt.xlabel("Номер на улица")
plt.show()



```
